```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f1(x):
    return (x-2)*x*(x+2)

def f2(x):
    return 3*(x-1)*(x-2)

z = np.linspace(-3,3,20)

y1 = f1(z)
y2 = f2(z)

print(y1)
print(y2)

plt.plot(x,y1, 'bo-', label = 'f1')
plt.plot(x,y2, 'r:', label = 'f2')

plt.legend(loc = 'upper left')
plt.title('My first Python graphs!!')
plt.xlabel('x')
plt.ylabel('y')

plt.xlim(-2,2)
plt.ylim(-20,20)
plt.grid(True)
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f2(x,w):
    return (x-w)*x*(x+2)

plt.figure(figsize=(10,3))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

x = np.linspace(-3, 3, 50)
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.title(i+1)
    plt.plot(x, f2(x,i), 'k')
    plt.ylim(-20, 20)
    plt.grid(True)
plt.show()    
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f3(x0, x1):
    r = 2*x0**2+x1**2
    return r*np.exp(-r)

n_point = 30
x = np.linspace(-2, 2 , n_point)
y = np.linspace(-2, 2 , n_point)
z = np.zeros((n_point, n_point))

for m in range(len(x)):
    for n in range(len(y)):
        z[m,n] = f3(x[m], y[n])
    
plt.figure(figsize = (3.5,3))

plt.gray()
plt.pcolor(z)
plt.colorbar()
plt.show()


plt.contour(x,y,z, 5)
plt.show()


from mpl_toolkits.mplot3d import Axes3D

xx, yy = np.meshgrid(x,y)

plt.figure(figsize=(5,4))

ax = plt.subplot(1,1,1, projection = '3d')
ax.plot_surface(xx,yy,z,rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')
ax.set_zticks((0,0.2))
ax.view_init(60,45)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def make_circle(c_x,c_y,r):
    theta = np.linspace(0, 2*3.14, 100) # (0 <= theta <= 2 * pi)
    x = r * np.cos(theta) # X  * cos()
    y = r * np.sin(theta) # y  * sin()   
    return plt.plot(x+c_x,y+c_y) # (x + x , y + y )

center_x = 5
center_y = 10
radius = 10

make_circle(center_x, center_y, radius)

plt.grid()
plt.title('2016146044_ChoYeongheum')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-6,16)
plt.ylim(-1,21)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def make_square_pulse(t):
    x = np.linspace(0,5,100)
    y = np.abs(x-np.round(x))<=(t/2)
    return plt.plot(x,y)

tau = 0.3

make_square_pulse(tau)


plt.grid()
plt.title('2016146044_ChoYeongheum')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-0.1, 5.1)
plt.ylim(-0.05,1.05)
plt.show()
```
